```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

## Dirty Water (exercise solution)
# CLEAN OR DIRTY WATER? HOW DIRTY?

#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from operator import truediv


#%%
# load image
img = (plt.imread("datawater/0_simple.jpg"))                  #image in dataWater/ directory
plt.imshow(img)

#%%
# get the R G B channels
r = np.uint(img[:,:,0])
g = np.uint(img[:,:,1])
b = np.uint(img[:,:,2])
#convert to grayscale image
I = (r+g+b)/3

## display the grayscale water image
fig = plt.figure(10)
fig.suptitle("Figure 10")
plt.imshow(I, cmap="gray")
plt.title("Dirty water image")
plt.show()
```

```python
#%%
## Calculate an approximate mean image in order to do light balancing.
# II.	Preprocess filter the images with low pass filters or with high pass filters.
# Lowpass filter with a large uniform kernel
dm = 300
filterMN = np.ones((dm,dm))
filterMN = filterMN/dm/dm
Imn = cv2.filter2D(I, -1,filterMN)

# display the mean image
fig = plt.figure(20)
fig.suptitle("Figure 20")
plt.imshow(Imn, cmap="gray")
plt.title("Mean image")
plt.show()
```

```python
#%%
## remove the mean image
# II.	Preprocess filter the images with low pass filters or with high pass filters.
Imnr = I - Imn
# and rescale - 0 to 255
Imnr = Imnr-np.nanmin(Imnr)
Imnr = (Imnr/np.nanmax(Imnr))*255
# display the mean removed image
fig = plt.figure(30)
fig.suptitle("Figure 30")
plt.imshow(Imnr, cmap="gray")
plt.title("Mean removed image")
plt.show()
```

```python
#%%
## Loop over several Threshold setting
# III.	Convert to Binary images with different thresholds.

count  = 0

for thresh in [200,130, 170]:
    count = count + 1

    # IV.	Isolate the particles. Apply logical filtering
    # (i.e. morphological filers) in order to clean up the image
    # and to approximately isolate each particle as a single
    # “connected component”.

    # Assumes particles are dark on light background.
    # For regions below threshold set flag 1 for each pixels that is below
    # threshold.
    BW = (Imnr<thresh).astype(np.uint8)

    fig = plt.figure((40+count))
    fig.suptitle("Figure "+ str(40+count))
    plt.imshow(BW, cmap="gray")
    titl = ("Threhsold image. Thresh = " + str(thresh))
    plt.title(titl)
    plt.show()

    ## removing of small the connected components
    #open
    #Remove small spots having a radius less than 10 pixels by opening it with
    #a disk-shaped structuring element having a 10 pixel radius.
    se = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
    BWafterOpening = cv2.morphologyEx(BW, cv2.MORPH_OPEN, se)

    fig = plt.figure((50+count))
    fig.suptitle("Figure "+ str(50+count))
    plt.imshow(BWafterOpening, cmap="gray")
    titl = ("Threhsold image, small removed. Thresh = " + str(thresh))
    plt.title(titl)
    plt.show()


    # V.	Count the particles and determine their properties (size, shape, etc).
    # a.	Apply image labeling routines
    # b.	Visualize histograms of several properties or property combinations.
    # c.	Can you classify different types of particles?

    ## labelling the connected components
    # label the connected components
    # L = bwlabel(BW)
    ret, L = cv2.connectedComponents(BWafterOpening, connectivity=8)
    fig = plt.figure((60+count))
    fig.suptitle("Figure "+ str(60+count))
    plt.imshow(L)
    titl = ("TConnected components, labeled image. Thresh = " + str(thresh))
    plt.title(titl)
    plt.show()
```

```python


    # VI.	Quantify distributions of particle properties for different
    # processing parameters.

    ## properties of the connected components
    connBW, dum = cv2.findContours(BW, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    connBWopening, dum = cv2.findContours(BWafterOpening.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    areaBW = []
    areaBWopening = []
    perimeterBW = []
    perimeterBWopening = []
    for c in connBW:
        area = cv2.contourArea(c)
        if (area > 10 and area < 50000):
            areaBW.append(area)
            perimeterBW.append(cv2.arcLength(c, True))
    for c in connBWopening:
        area = cv2.contourArea(c)
        if (area > 10 and area < 50000):
            areaBWopening.append(area)
            perimeterBWopening.append(cv2.arcLength(c, True))


    ## plots of histograms of the connected components
    #fig = plt.figure((70+count))
    fig = plt.figure(70+count, constrained_layout=False)
    fig.suptitle(("Figure "+ str(70+count)))

    plt.subplot(3,2,1)
    plt.hist(areaBW, bins=10, density=False)
    titl = (" Hist. Area.  Thresh = " + str(thresh))
    plt.title(titl)
    plt.xlabel("Area")
    plt.ylabel("Count")

    plt.subplot(3,2,3)
    plt.hist(perimeterBW, bins=10, density=False)
    titl = (" Hist. Perimeter ")
    plt.title(titl)
    plt.xlabel("Perimeter")
    plt.ylabel("Count")

    plt.subplot(3,2,5)
    #plt.hist(areaBW/perimeterBW, bins=10, density=False)
    plt.hist(list(map(truediv, areaBW, perimeterBW)), bins=10, density=False)
    titl = (" Hist. A/P ")
    plt.title(titl)
    plt.xlabel("Area / Perimeter")
    plt.ylabel("Count")

    plt.subplot(3,2,2)
    plt.hist(areaBWopening, bins=10, density=False)
    titl = (" Hist. Area. Small removed. Thresh =  " + str(thresh))
    plt.title(titl)
    plt.xlabel("Area")
    plt.ylabel("Count")

    plt.subplot(3,2,4)
    plt.hist(perimeterBWopening, bins=10, density=False)
    titl = (" Hist. Perimeter ")
    plt.title(titl)
    plt.xlabel("Perimeter")
    plt.ylabel("Count")

    plt.subplot(3,2,6)
    #plt.hist(areaBWopening/perimeterBWopening, bins=10, density=False)
    plt.hist(list(map(truediv, areaBWopening, perimeterBWopening)), bins=10, density=False)
    titl = (" Hist. A/P ")
    plt.title(titl)
    plt.xlabel("Area / Perimeter")
    plt.ylabel("Count")

    plt.show()
```
